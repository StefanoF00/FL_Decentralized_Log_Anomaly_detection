```python
Q = 5.0 # dVdx < -Q

# Function to enforce hard constraints
def hard_constraint(x, y):
    '''
    Enforces output y=[V,u] to have V(0)=0 and V_dot(0)=0
    
    x_example = [0; 2; 0]          # Shape (3, 1)
    y_example = [4, 5; 6, 7; 8, 9] # Shape (3, 2)
    x_example * y_example = [0, 0; 12, 14; 0, 0]
    '''
    V = x * y[:, 0:1]  # Element-wise multiplication to apply the constraint
    u = y[:, 1:2]
    # Concatenate V and u along the second dimension
    constrained_y = torch.cat((V, u), dim=1)
    return constrained_y

# Define the ODE function for the Lyapunov constraints
def integrator_ode(pred, x):
    '''
    System considered: INTEGRATOR
    - system: dxdt = u
    - input:  u to be learned  
    '''
    dxdt =  pred[:,1] #+ 0.6*sin(x[:,1])
    dVdx = jacobian(pred, x, i=0,j=0) 
    return dVdx*dxdt
```

```python
def learned_control(state, model):
    #theta_normalized, omega_normalized = coordinate_transformation(theta, omega)
    input_tensor = torch.tensor(np.array([state]), dtype=torch.float32).to('cuda:0')
    control = model(input_tensor)[:, 1].item()
    return control

def disturb(param, x, t):
    # Example disturbance function
    return param * np.sin(t)
    
def integrator_dynamics(t, state, model):
    u = learned_control(state, model)
    dxdt = u
    return [dxdt]

initial_state = [5.0]

# Simulation time
t_span = (0, 10)
t_eval = np.linspace(t_span[0], t_span[1], 1000)

if load_model is True:
    model_path = f'output/Integrator_PINN/Integrator_PINN_{num_training}/model/model.pt'
    loaded_model = torch.load(model_path)
    loaded_model.eval()
    solution = solve_ivp(integrator_dynamics, t_span, initial_state, args=(loaded_model,), t_eval=t_eval, method='RK23')
else:
    solution = solve_ivp(integrator_dynamics, t_span, initial_state, args=(integrator_model,), t_eval=t_eval, method='RK23')

state_values = solution.y[0]
time = solution.t

plt.figure(figsize=(10, 5))
plt.plot(time, state_values, label=f'$x(0)=${(initial_state[0]):.4f}; Final value={state_values[-1]:.4f}')
plt.xlabel('Time [s]')
plt.ylabel('$x(t)$')
plt.legend()
plt.title('Time evolution of $x(t)$')
plt.show()
```

# Consensus for integrator

Define the network

```python
N = 5
def create_specific_leader_follower_graph(N):
    # Number of nodes (0 to 5)
    
    # Initialize the adjacency matrix with zeros
    A = np.zeros((N, N), dtype=int)
    
    # Manually define the edges according to the given graph
    edges = [
        (0, 1), (0, 3), (0, 4),
        (1, 2), 
        (3, 4), (3, 5), 
        (4, 5)
    ]
    
    # Populate the adjacency matrix
    for (u, v) in edges:
        A[u][v] = 1
        A[v][u] = 1  # Since the graph is undirected
    D = np.diag(np.sum(A, axis=1))
    L = D - A
    return A, D, L

seed = 42
A, D, L = create_specific_undirected_graph()
#A, D, L = create_random_undirected_graph(N, p=0.3)
#A, D, L = create_connected_undirected_graph(N, extra_edges=0)

plot_eigenvalues(L)
plot_network(A)
```

```python
# LOAD THE MODEL
import time
load_model = True
if load_model is True:
    num_training = "03"
    model_path = f'output/Integrator_PINN/Integrator_PINN_{num_training}/model/model.pt'
    model = torch.load(model_path)
    model.eval()
    model_path_2 = f'output/Integrator_PINN_check_alpha/Integrator_PINN_check_alpha_08/model/model.pt'
    model_2 = torch.load(model_path_2)
    model_2.eval()
else:
    model = integrator_model
    model_2 = integrator_model_2

# DEFINE PARAMETERS
M = 4
k1 = 3.5     # Initial gain for v_i
k2 = 1.4     # Initial gain for tanh term
delta = 5.4
alpha = 21/25  # Exponent for v_i (radio of pos ODD numbers)
delta1 = 4  #delta - k2 # Saturation threshold
Tt = 100    # Number of time steps
dt = 0.01    # Time step size
dist = 0.0
c= 1
# Initialize the leader state
state_ranges = generate_random_state_ranges(M, overall_range=(-1, 1))
initial_states = initialize_integrator_states(N,M,state_ranges)

for m in range(M):
    states_for_m = initial_states[:, m]
    print(f"Initialization Weights {m+1}: {states_for_m};\t expected consensus at: {np.mean(states_for_m)}")

# Matrix of NxM time intervals defined by lemma 2 (one for each initial state) 
T_m = compute_T(model_2, initial_states, alpha, c, 0.9)

# Vector of M time intervals (m-th elem is the biggest one for the m-th consensus protocol)
T = np.zeros(M)
for m in range(M):
    T[m] = int(np.max(T_m[m])+1)

# M consensuses are simulated, each one for T[m] time intervals
start_data = time.perf_counter()
final_state, trajectory, time_values = simulate_network(N, M, T, dt, A, k1, k2, alpha, delta1, initial_states, model, dist)
end_data = time.perf_counter()
training_time = (end_data - start_data)
print(f'Consensus performed in {training_time:.4f}s!')

data = {
    'trajectory': trajectory,
    'time_values': time_values,
    'final_state':final_state
}
savemat(f'{figures_path}consensus_data.mat', data)
# Plot of the results
plot_convergence_M_states(trajectory, time_values)
print(final_state)
#plot_integrator_M_controls(trajectory, time_values)
#plot_integrator_control(trajectory, time_values)
```
